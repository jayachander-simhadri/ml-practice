```python
#custome iteration example
class Count:
    def __init__(self, n):
        self.n = n
        self.cur = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.cur > self.n:
            raise StopIteration
        val = self.cur
        self.cur += 1
        return val
```

```python
c = Count(3)

for x in c: # thhis for loop only knows to call it.__iter__() and it.__next__(), if we not defines error will be thrown
    print(x)
```

```python
#zip functionality
names = ["A", "B", "C"]
scores = [90, 85]

res = zip(names, scores)

print(list(res))  # same as generator, executed when needed once it is done it is exhausted.
# [('A',90), ('B',85), ('C',88)]
```

```python
#map use : will apply the evaluation to each item in the iteratble map(function,iterable)

lst = [1,2,3,4,5]
t = (3,4,5)

m = map(lambda x : x**0.5, lst)
print(list(m))

f = filter(lambda x : x%3==0,lst)
print(list(f))

z=zip(lst,t)
print(list(z))
#below three statements show that all the above iterators ended after print i.e exhausted so 
# we again used print it will return [] same as generator funtionality from python 3
print(list(m))
print(list(f))
print(list(z))
```

```python
#list builder function in terms of generator

def gen_lst(x):
    for i in range(x):
        yield i*i

g = gen_lst(5)

for x in g:
    print(x)
print(list(g))
```
