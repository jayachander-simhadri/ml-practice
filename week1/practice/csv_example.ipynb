```python
import csv

rows = [
    ['name','age','city'],
    ['jc','22','vij'],
    ['varun','23','hyd'],
    ['katraj','23','elr']
]

with open('sample.csv','w+',newline="") as f:
    writer  = csv.writer(f)

    for row in rows:
        writer.writerow(row)
    f.seek(0)
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
with open('sample.csv','r',newline="") as f:
    reader= csv.DictReader(f)
    for row in reader:
        print(row)
```

```python
with open('sample.csv','r')as f:
    reader = csv.reader(f)
    for row in reader:
        clean_row = [cell.strip() for cell in row if cell.strip()!=""]
        print(clean_row)
```

```python
with open('sample.csv','r')as f:
    reader = csv.DictReader(f)
    for row in reader:
        clean1_row={
            key.strip():value.strip()
            for key,value in row.items()
            if key and value and value.strip()!=''#here 'key and value' used because there if new coloum there for any row but header is 'None'
            #it will throw attribute error so to skip it we also validating key to exorcist that column in that row
        }
        print(clean1_row)
```
