```python
#loops example
def chk_range(a,b,c=1):
    chk_list = []
    if not all(isinstance(x,int) for x in [a,b,c]):
        return "all numbers must be integers"
    for i in range(a,b,c):
        print(i)
#below is example to count number of digits in number
def count(num):
    count =0
    if num == 0:
        return 1
    if not isinstance(num,int):
        return "enter valid number"
    num = abs(num)
    while num>0:
        count+=1
        num = num//10
    return count
print(count(987656))
print(chk_range(2,10,2))
```

```python
#find key with max value in dict
from collections import defaultdict
from operator import itemgetter

dict1 = defaultdict(int)
ex_list = [1,2,3,3,4,1,2,1,1,0,0,0,0,0]
for i in ex_list:
    dict1[i]+=1
#temp = sorted(ex_list)
temp = list(sorted(dict1.items(), key=itemgetter(1),reverse=False))
print(temp)
max_key = max(dict1,key = dict1.get)
print(max_key)
```

```python
#sum and average
def sum_and_average(nums):
    total = sum(nums)
    avg = total/len(nums)
    return total,avg
nums = [10,5,6,2.4,0]
print(type(sum_and_average(nums)))

def isprime(num):
    if num <= 1:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            return False
    return True
print(isprime(7))
print(isprime(10))
    
```
