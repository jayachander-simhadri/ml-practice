```python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    filename= "ex.log",
    filemode= "a"
)

logging.debug("Debug message")
logging.info("program started")
logging.error("something error occured")
logging.warning("crashed")
logging.critical("application crashed critical!!")
```

```python
try:
    x = 10/0
except Exception:
    logging.exception("error")
```

```python
logger = logging.getLogger(__name__)
logger.info("checking the new Logger")
```

```python
def login(username,password):
    if username=="admin" and password=="1234":
        logger.info(f"{username} logged in successfully")
    else:
        logger.warning(f"failed login attempt for {username}")

login("admin","1234")
login("admin","weong")
```

```python
rows = [
    {"name":"jaya","age":22},
    {"name":"kc","age":""}
]

for row in rows:
    if not row['age']:
        logger.error(f"invalid age field for {row['name']}")
    else:
        logger.info(f"processing age for {row['name']}")
```
